## Series: It is defined as a 1D array that is capable of storing various data types
- import pandas as pd
- a= pd.Series()
- print(a)

```python
import pandas as pd
```

```python
x=[2,3,4,5,6,9]
var=pd.Series(x)
print(var)
print(type(var))
print(var[3])
```

## To change index, datatype, give name to your data

```python
x=[2,3,4,5,6,9]
var=pd.Series(x,index=['a','b','c','d','e','f'],dtype='float',name='python')

print(var)
print(type(var))
```

```python
dic={                                       #when we use mixed data datatype is shown object
    'name':['python','C','C++','java'],
    'pop':[12,13,14,15],
    'rank':[1,2,3,4]
}
var1=pd.Series(dic)
print(var1)
```

```python
s= pd.Series(12,index=[1,2,3,4,5,6,7])
print(s)
print(type(s))
```

```python
s1= pd.Series(12,index=[1,2,3,4,5,6,7])
s2= pd.Series(12,index=[1,2,3,4])
print(s1)
print(s2)
```

```python
s1= pd.Series(12,index=[1,2,3,4,5,6,7])
s2= pd.Series(12,index=[1,2,3,4])
print(s1+s2)
```

## DataFrame-2D array

```python
l=[1,2,3,4,5]
va=pd.DataFrame(l)
print(va)
print(type(va))
```

```python
di={
    'a':[1,2,3,4,5],
    's':[6,7,8,9,10]
}
va1=pd.DataFrame(di)
print(va1)
print(type(va1))
```

```python
di1={
    'a':[1,2,3,4,5],
    's':[6,7,8,9,10],
    'd':[11,12,13,14,15],
    'e':[1,2,3,4,5]
}
va2=pd.DataFrame(di1,columns=['a','e'],index=['A','B','C','D','E'])  #INDEX CHANDES INDEX 
print(va2)
print(type(va2))
```

```python
di1={                              #dict
    'a':[1,2,3,14,5],
    's':[6,7,8,9,10],
    'd':[11,12,13,14,15],
    'e':[1,2,3,4,5]
}
va2=pd.DataFrame(di1)
print(va2)
print()
print(va2['a'][3])
```

```python
list_1 = [[1,2,3,4,5],[11,12,13,14,15]]   #list
var2 = pd.DataFrame(list_1)
print(type(var2))
print(var2)
```

```python
sr={'s':pd.Series([1,2,3,4]),'r':pd.Series([1,2,3,4])}
var3 = pd.DataFrame(sr)
print(type(var3))
print(var3)
```

## Arithmetic operations 

```python
va2={
    'A':[1,2,3,4],
    'B':[5,6,7,8]
}
va3= pd.DataFrame(va2)
print(va3)
```

```python
va3['C'] = va3['A']+va3['B']  #sub,mul,div
va3
```

```python
va4=pd.DataFrame({
    'A':[10,20,30,40],
    'B':[15,16,17,18]
})
va4
```

```python
va4['python']=va4['A']<=20   #data less then 20 gives True
va4['python_1']=va4['B'] >=16
va4 
```
