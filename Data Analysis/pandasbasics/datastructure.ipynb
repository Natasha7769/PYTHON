## Merging & Concat

# merge

```python
num = pd.DataFrame({'A':[1,2,3,4,5],'B':[11,12,13,14,15]})
num1 = pd.DataFrame({'A':[1,2,3,4,5],'C':[21,22,23,24,25]})
pd.merge(num,num1,on='A')
```

```python
num = pd.DataFrame({'A':[1,2,3,4,5],'B':[11,12,13,14,15]})
num1 = pd.DataFrame({'A':[1,2,3,4,6],'C':[21,22,23,24,25]})  #common data will be displayed
pd.merge(num,num1,on='A')
```

```python
num = pd.DataFrame({'A':[1,2,3,4,5],'B':[11,12,13,14,15]})
num1 = pd.DataFrame({'A':[1,2,3,4,6],'C':[21,22,23,24,25]})
pd.merge(num,num1,how = 'left') #right #outer #inner
```

```python
num = pd.DataFrame({'A':[1,2,3,4,5],'B':[11,12,13,14,15]})
num1 = pd.DataFrame({'A':[1,2,3,4,6],'C':[21,22,23,24,25]})
pd.merge(num,num1,how= 'outer',indicator= True)
```

```python
num = pd.DataFrame({'A':[1,2,3,4,5],'B':[11,12,13,14,15]})
num1 = pd.DataFrame({'A':[1,2,3,4,6],'B':[21,22,23,24,25]})
pd.merge(num,num1,left_index=True,right_index=True)
```

```python
num = pd.DataFrame({'A':[1,2,3,4,5],'B':[11,12,13,14,15]})
num1 = pd.DataFrame({'A':[1,2,3,4,6],'B':[21,22,23,24,25]})
pd.merge(num,num1,left_index=True,right_index=True,suffixes=('name','id'))
```

# Concat

```python
sr1 = pd.Series([1,2,3,4])
sr2 = pd.Series([11,21,31,41])
pd.concat([sr1,sr2])
```

```python
num = pd.DataFrame({'A':[1,2,3,4,5],'B':[11,12,13,14,15]})
num1 = pd.DataFrame({'A':[1,2,3,4,6],'B':[21,22,23,24,25]})
pd.concat([num,num1])
```

```python
num = pd.DataFrame({'A':[1,2,3,4,5],'B':[11,12,13,14,15]})
num1 = pd.DataFrame({'A':[1,2,3,4,6],'B':[21,22,23,24,25]})
pd.concat([num,num1],axis=1) #axis=0
```

```python
num = pd.DataFrame({'A':[1,2,3,5,7],'B':[11,12,13,14,15]})
num1 = pd.DataFrame({'A':[1,2,4,6],'C':[21,22,23,24]})
pd.concat([num,num1],axis=1)
```

```python
num = pd.DataFrame({'A':[1,2,3,5,7],'B':[11,12,13,14,15]})
num1 = pd.DataFrame({'A':[1,2,4,6],'C':[21,22,23,24]})
pd.concat([num,num1],axis=1,join='inner')
```

```python
num = pd.DataFrame({'A':[1,2,3,4,5],'B':[11,12,13,14,15]})
num1 = pd.DataFrame({'A':[1,2,3,4,6],'B':[21,22,23,24,25]})
pd.concat([num,num1],axis=1,keys=['d1','d2'])  #axis=0
```

```python
num = pd.DataFrame({'A':[1,2,3,4,5]})
num1 = pd.DataFrame({'A':[1,2,3,4,6],'B':[21,22,23,24,25]})
pd.concat([num,num1])
```
