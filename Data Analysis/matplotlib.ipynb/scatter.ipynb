```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
plt.scatter(day,no,color='r')
plt.title('scatter plot',fontsize=15)
plt.xlabel('day',fontsize=15)
plt.ylabel('no',fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=['r','y','g','b','r','g','r']
size=[100,200,400,500,100,100,70]
plt.scatter(day,no,c=colors,s=size,alpha=0.5)
plt.title('scatter plot',fontsize=15)
plt.xlabel('day',fontsize=15)
plt.ylabel('no',fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=['r','y','g','b','r','g','r']
size=[100,200,400,500,100,100,70]
plt.scatter(day,no,c=colors,s=size,marker='*',edgecolor='g',linewidth=2)
plt.title('scatter plot',fontsize=15)
plt.xlabel('day',fontsize=15)
plt.ylabel('no',fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=[10,40,30,20,50,20,30]
size=[100,200,400,500,100,100,70]
plt.scatter(day,no,c=colors,s=size,cmap='turbo',alpha=0.5)
t=plt.colorbar()
t.set_label('colorbar',fontsize=15)
plt.title('scatter plot',fontsize=15)
plt.xlabel('day',fontsize=15)
plt.ylabel('no',fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
no2=[3,2,4,5,1,6,2]
colors=[10,40,30,20,50,20,30]
size=[100,200,400,500,100,100,70]
plt.scatter(day,no,c=colors,s=size,cmap='turbo',alpha=0.5)
plt.scatter(day,no2,color='r',s=size,alpha=0.5)
t=plt.colorbar()
t.set_label('colorbar',fontsize=15)
plt.title('scatter plot',fontsize=15)
plt.xlabel('day',fontsize=15)
plt.ylabel('no',fontsize=15)
plt.show()
```
