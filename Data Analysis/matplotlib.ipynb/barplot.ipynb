## graphs for visualization
- line- for understanding the trend
- area- for understanding the trend
- bar- for comparing the values (discrete values)
- pie- for comparing the values (percentage based)
- boxplot- for understanding the outliers and quartiles (statistical analysis)
- histogram- for understanding the distribution of data( statistical analysis)
- scatter- for understanding the relationship b/w two or more variables
- heatmap- 2d histogram

```python
x=['python','C','C++','Java']
y=[85,70,60,82]
plt.xlabel('languages',fontsize=20)
plt.ylabel('Numbers',fontsize=20)
plt.title('data',fontsize=20)
c=['y','b','m','g']
plt.bar(x,y,width=0.4,color=c,align='edge')  #align=center by default
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=['python','C','C++','Java']
y=[85,70,60,82]
z=[20,30,40,50]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel('languages',fontsize=15)
plt.ylabel('Numbers',fontsize=15)
plt.title('data',fontsize=15)

plt.bar(p,y,width,color='r',label='POPULARITY')  
plt.bar(p1,z,width,color='y',label='POPULARITY1')

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
x=['python','C','C++','Java']
y=[85,70,60,82]
z=[20,30,40,50]

plt.xlabel('languages',fontsize=15)
plt.ylabel('Numbers',fontsize=15)
plt.title('data',fontsize=15)

plt.barh(x,y,color='r',label='POPULARITY')  
plt.barh(x,z,color='y',label='POPULARITY1')

plt.legend()
plt.show()
```
