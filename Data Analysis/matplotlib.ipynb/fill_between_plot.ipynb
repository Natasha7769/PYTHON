```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [2,4,6,8,10]
y= [5,2,6,4,8]
plt.plot(x,y,color='r')
plt.fill_between(x,y,color='g')
plt.title('python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
x = [2,4,6,8,10]
y= [5,2,6,4,8]
plt.plot(x,y,color='r')
plt.fill_between(x=[4,8],y1=6,y2=4)
plt.title('python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
import numpy as np
x = np.array([2,4,6,8,10])
y= np.array([5,2,6,4,8])
plt.plot(x,y,color='r')
plt.fill_between(x,y,color='g',where=(x>=2) & (x<=4),alpha=0.5)  #alpha=convert in light color
plt.title('python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```
