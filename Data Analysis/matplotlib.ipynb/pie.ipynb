```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40]
y=['c','c++','java','python']
plt.pie(x,labels=y)
plt.show()
```

```python
x=[10,20,30,40]
y=['c','c++','java','python']
ex=[0.4,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex)
plt.show()
```

```python
x=[10,20,30,40]
y=['c','c++','java','python']
ex=[0.4,0.0,0.0,0.0]
c=['r','b','g','y']
plt.pie(x,labels=y,explode=ex,colors=c)
plt.show()
```

```python
x=[10,20,30,40]
y=['c','c++','java','python']
ex=[0.4,0.0,0.0,0.0]
c=['r','b','g','y']
plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.2f%%')
plt.show()
```

```python
x=[10,20,30,40]
y=['c','c++','java','python']
ex=[0.4,0.0,0.0,0.0]
c=['r','b','g','y']
plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.2f%%',shadow=True)
plt.show()
```

```python
x=[10,20,30,40]
y=['c','c++','java','python']
ex=[0.4,0.0,0.0,0.0]
c=['r','b','g','y']
plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.2f%%',radius=2)
plt.show()
```

```python
x=[10,20,30,40]
y=['c','c++','java','python']
ex=[0.4,0.0,0.0,0.0]
c=['r','b','g','y']
plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.2f%%',labeldistance=2)
plt.show()
```

```python
x=[10,20,30,40]
y=['c','c++','java','python']
ex=[0.4,0.0,0.0,0.0]
c=['r','b','g','y']
plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.2f%%',startangle=90) #you can rotate your chart
plt.show()
```

```python
x=[10,20,30,40]
y=['c','c++','java','python']
ex=[0.4,0.0,0.0,0.0]
c=['r','b','g','y']
plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.2f%%',textprops={'fontsize':15})
plt.show()
```

```python
x=[10,20,30,40]
y=['c','c++','java','python']
ex=[0.4,0.0,0.0,0.0]
c=['r','b','g','y']
plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.2f%%',counterclock=False)
plt.show()
```

```python
x=[10,20,30,40]
y=['c','c++','java','python']
ex=[0.4,0.0,0.0,0.0]
c=['r','b','g','y']
plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.2f%%',shadow=True,radius=1.3,labeldistance=1.1,
        startangle=0,textprops={'fontsize':15},counterclock=False,wedgeprops={'linewidth':4,'edgecolor':'m'},rotatelabels=True)
plt.show()
```

```python
x=[10,20,30,40]
y=['c','c++','java','python']
ex=[0.4,0.0,0.0,0.0]
c=['r','b','g','y']
plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.2f%%',counterclock=False)
plt.title('Languages')
plt.legend(loc=2) #location=1 or 2 or 3 or 4
plt.show()
```

```python
x=[10,20,30,40]
x1=[40,30,20,10]
y=['c','c++','java','python']
c=['r','b','g','y']
plt.pie(x,labels=y,radius=1.5)
plt.pie(x1,radius=1,colors=c)

plt.show()
```

```python
x=[10,20,30,40]    #method 1 donut 
x1=[40,30,20,10]
y=['c','c++','java','python']
c=['r','b','g','y']
plt.pie(x,labels=y,radius=1.5)
plt.pie([1],colors='w')

plt.show()
```

```python
x=[10,20,30,40]    #method 2
x1=[40,30,20,10]
y=['c','c++','java','python']
c=['r','b','g','y']
plt.pie(x,labels=y,radius=1.5)
cr=plt.Circle(xy=(0,0),radius=1,facecolor='w')
plt.gca().add_artist(cr)


plt.show()
```
