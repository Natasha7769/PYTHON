```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
area=[2,4,3,6,5]

plt.stackplot(x,area)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,4,3,6,5]
area2=[1,4,2,6,2]
area3=[4,2,3,1,5]

plt.stackplot(x,area1,area2,area3)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,4,3,6,5]
area2=[1,4,2,6,2]
area3=[4,2,3,1,5]
l=['area1','area2','area3']

plt.stackplot(x,area1,area2,area3,labels=l)
plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,4,3,6,5]
area2=[1,4,2,6,2]
area3=[4,2,3,1,5]
l=['area1','area2','area3']

plt.stackplot(x,area1,area2,area3,labels=l)
plt.legend(loc=2)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,4,3,6,5]
area2=[1,4,2,6,2]
area3=[4,2,3,1,5]
l=['area1','area2','area3']

plt.stackplot(x,area1,area2,area3,labels=l,colors=['r','g','m'])
plt.legend(loc=2)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,4,3,6,5]
area2=[1,4,2,6,2]
area3=[4,2,3,1,5]
l=['area1','area2','area3']

plt.stackplot(x,area1,area2,area3,labels=l,colors=['r','g','m'],baseline='sym')  #zero,wiggle
plt.title('python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.legend(loc=2)
plt.show()
```
