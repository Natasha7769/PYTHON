### syntax-plot()
- <matplotlib.pyplot>.plot(x-axix, y-axis ,color ,linewidth ,linestyle or ls ,marker ,markersize ,markeredgecolor)
- linewidth - float value
- linestule or ls- solid lines,dashed line
- marker- 'o','*', '+' etc
- markersize- float value
- markeredgecolor- color identified by pyplot

```python
import matplotlib.pyplot as plt
x = [2,4,6,8,10]
y= [5,2,6,4,8]
plt.plot(x,y,color='r')
plt.show()
```

```python
y= [5,2,6,4,8] # only one value it will take value at y axis and take x axis value as 0,1,2,3,4...
plt.plot(y,color='r')
plt.show()
```

```python
x = [2,4,6,8,10]                                            #Linessytle attributes
y= [5,2,6,4,8]                                                #solid or -[defaule]
plt.plot(x,y,color='r',linewidth=10,linestyle='dashdot')      # dashed or _     
plt.show()                                                   # dotted or :   # dashdot or -
```

```python
x = [2,4,6,8,10]
y= [5,2,6,4,8]
plt.plot(x,y,color='r',marker='*',markeredgecolor='b',markersize=10)
plt.show()
```

```python
x = [5,10,15,20,25]
y= [12,4,15,18,12]
plt.plot(x,marker='*') # y axis value
plt.plot(y,marker='+')  # y axis value
plt.show()
```
