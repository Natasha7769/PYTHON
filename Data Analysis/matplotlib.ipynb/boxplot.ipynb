```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x)
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x,notch=True)  #False
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x,vert=False)
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x,widths=0.5)
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x,labels=['python'],patch_artist=True)  #data given in label always in list
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x,showmeans=True)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120]  #120 is outlier
plt.boxplot(x)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120,130]  
plt.boxplot(x)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120,130,300]  
plt.boxplot(x)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120]  # whis join outlier
plt.boxplot(x,labels=['python'],showmeans=True,whis=2)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120,130,300]  
plt.boxplot(x,sym='g+')  #to hide outlier sym=" "
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120]  #max and min line called cap
plt.boxplot(x,labels=['python'],showmeans=True,boxprops=dict(color='r'),capprops=dict(color='g'),whiskerprops=dict(color='r'),flierprops=dict(markeredgecolor='y'))
plt.show()
```

```python
x=[10,20,30,40,50,60,70]  #max and min line called cap
x1=[10,20,40,50,30,60,90]
y=[x,x1]
plt.boxplot(y,labels=['pyhton','c++'],showmeans=True,boxprops=dict(color='r'),capprops=dict(color='g'),whiskerprops=dict(color='r'),flierprops=dict(markeredgecolor='y'))

plt.show()
```
